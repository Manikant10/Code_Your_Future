```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a synthetic dataset
# Let's assume we have the following data for house sizes (in square feet) and their prices (in thousands)
# For example: sizes = [1500, 1600, 1700, 1800, 1900] and prices = [300, 320, 340, 360, 380]
sizes = np.array([1500, 1600, 1700, 1800, 1900])
prices = np.array([300, 320, 340, 360, 380])

# Step 2: Calculate the slope (m) and intercept (b) using the least squares formula
def linear_regression(X, y):
    n = len(X)
    X_mean = np.mean(X)
    y_mean = np.mean(y)
    
    # Calculate the slope (m)
    numerator = np.sum((X - X_mean) * (y - y_mean))
    denominator = np.sum((X - X_mean) ** 2)
    m = numerator / denominator
    
    # Calculate the intercept (b)
    b = y_mean - m * X_mean
    
    return m, b

# Step 3: Train the model
m, b = linear_regression(sizes, prices)

# Step 4: Make predictions
def predict(X, m, b):
    return m * X + b

predicted_prices = predict(sizes, m, b)

# Step 5: Visualize the results
plt.scatter(sizes, prices, color='blue', label='Actual Prices')
plt.plot(sizes, predicted_prices, color='red', label='Regression Line')
plt.xlabel('Size of House (sq ft)')
plt.ylabel('Price (in thousands)')
plt.title('House Prices Prediction using Linear Regression')
plt.legend()
plt.show()

# Print the slope and intercept
print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")
```
